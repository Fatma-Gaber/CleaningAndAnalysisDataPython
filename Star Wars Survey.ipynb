### import Pandas Library to Read The data Which has diffrent encoding

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
#Explore first 10 row of Data 
star_wars.head(10)
```

```python
#Explore Columns of Data
star_wars.columns
```

```python
# Find unique value in the following Column
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
# Find unique value in the following Column
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

### Cleaning and Mapping Yes\No Columns

```python
new={
    "Yes":True,
    "No":False
}
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?']=star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(new)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']=star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(new)
```

```python
star_wars.head()
```

### Cleaning And Mapping Checkbox Columns

```python
import numpy as np
```

```python
newmapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
     np.nan: False
}
```

```python
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(newmapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

### Cleaning The Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })

star_wars.head()
```

### Compute the average ranking of each movie

```python
MeanRanking=star_wars[star_wars.columns[9:15]].mean()
MeanRanking
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### Visualize Mean Ranking For each movie 

```python
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

Conclusion : From Visualiztion We Noticed that "original" movies are highly rated than the newer

## Sum number of responded Who watch  Each movie And visualize it 

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

Conclusion: The original Movies Were seen by Respondents than The newer which explain why that the ranking for the original movies were higher 


### Identify mean number of Males And females who watched  the  movies  And Ranking them

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()
```

```python
plt.bar(range(6), males[males.columns[3:9]].sum())
plt.show()

plt.bar(range(6), females[females.columns[3:9]].sum())
plt.show()
```

More males watches movies 1-3, but males liked them less than females did that what ranking.
